```python
Unsupervised - Predicting patterns without specific target variable. 
New records(samples) can be assigned to existing clusters. 
1. Clustering - Groups samples with similar patterns. Clustering methods can be organized by modeling approaches like Centroids and hierarchial
* KMeans - remembers the centroid to each cluster and finds the nearest centroid to each new sample. 
Cross tabulation is used to evaluate the quality of clusters and informs the choice of how many clusters are needed. 
Clusters that are tight and close are determined as good. 
Inertia is used as a metric to find the spread of clusters. low inertia means clusters are tight. 
KMeans attempts to minimize inertia, when choosing clusters. 
Elbow plot is the most commonly used method to find the point where inertia begins to decrease. This determines the number of clusters being used. 

* Hierarchial and t-SNE - These are unsupervised techniques for visualization. 
* Hierarchial - Samples are arranged in hierarchial structure, that are visulaized as tree like diagram known as dendrogram. 
Agglomerative and Divisive clustering are two types of hierarchial clusters. 
1. Agglomerative - Two closest clusters merge together and the process continue until a single big cluster is left. 
2. Divisive - Single big cluster is divided until  a single sample of its own cluster os left. 
* Height on the dendrogram specifies maximum distance between merging clusters. 
Cluster labels can be extrated using fcluster. 

* tSNE - stands for t-distributed stochastic neighbor embedding. Maps samples from a high-dimensional space into 2D or 3D space. 
learning rates vary, but usually try values between 50-200
* PCA - Principle component analysis is a dimensionality reduction technique. It compresses high dimensional data to lower dimensional data. 
And it can also change the compressed data back to its original high dimensional. 
PCA speeds up the fitting of ML algorithms. 
```

```python
df.head(2)
```

```python
#Converting InvoiceDate to datetime
import datetime as dt
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['InvoiceDay'] = df['InvoiceDate'].apply(lambda x: dt.datetime(x.year, x.month, x.day))
pin_df=max(df['InvoiceDay'])+dt.timedelta(1)
pin_df
```

```python
#df["InvoiceDate"].min()
#df["InvoiceDate"].max()
```

```python
grouped_customerid=df.groupby('CustomerID').agg({'InvoiceDay':lambda x:(pin_df-x.max()).days,
                                  'InvoiceNo':'count','Quantity_price':'sum'})
grouped_customerid.sort_values(by=['Quantity_price'], inplace=True)
grouped_customerid.tail()
```

```python
#Targeting customers where 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
plt.title("Quantity_price")
sns.axes_style("dark")
sns.violinplot(y=df["Quantity_price"])
plt.show()
```

```python
#InvoiceDay	InvoiceNo	Quantity_price
grouped_customerid.tail()
```

```python
#Evaluating clustering
from sklearn.cluster import KMeans
wcss = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, init="k-means++", algorithm='full', max_iter=300)
    kmeans.fit(grouped_customerid.iloc[:,1:])
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(15,20))    
plt.grid()
plt.plot(range(1,10),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,10,1))
plt.ylabel("WCSS")
plt.show()
```

```python
#transforming features for better clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
scaler = StandardScaler()
kmeans = KMeans(n_clusters=3)

from sklearn.pipeline import make_pipeline
pipeline = make_pipeline(scaler, kmeans)
pipeline.fit(grouped_customerid.iloc[:, 1:])
labels = pipeline.predict(grouped_customerid.iloc[:, 1:])
```

```python
#Plotting scatter plot
x = grouped_customerid.iloc[:,1]
y = grouped_customerid.iloc[:,2]
plt.scatter(x, y, c=labels)
plt.show()
```

```python
centroids = kmeans.cluster_centers_
centroids_x = centroids[:, 0]
centroids_y = centroids[:, 1]
plt.scatter(centroids_x, centroids_y, marker='D', s=50)
plt.show()
```

```python
#spread of samples can be measured by inertia(lower is the better)
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(grouped_customerid.iloc[::])
print(model.inertia_)
```

5. Perform hierarchical clustering for customer segments. What is the ideal number of clusters? Which approach did you use to find the ideal number of clusters and why?

```python
#Crosstabulation is used for evaluating clusters. This measures the quality of clusters and determines the number of clusters. 
```

```python
grouped_customerid=df.groupby(['CustomerID']).agg({'InvoiceDay':lambda x:(pin_df-x.max()).days,
                                  'InvoiceNo':'count','Quantity_price':'sum'})
grouped_customerid.sort_values(by=['CustomerID'], inplace=True)
grouped_customerid.tail()
```

```python
#centroid - the distance between clusters equals the square of the Euclidean distance between the centroids of each cluster. 
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.preprocessing import normalize
normalized_group = normalize(grouped_customerid.iloc[:,1:])
mergings = linkage(grouped_customerid.iloc[:,1:], method='centroid')
from scipy.cluster.hierarchy import fcluster

labels = fcluster(mergings, 3, criterion='distance')
print(labels)
```

```python
#crosstabulation
ct = pd.crosstab(df['Quantity_price'], df['InvoiceDay'])
print(ct)
```

```python
dendrogram(mergings, labels=labels, leaf_rotation=90, leaf_font_size=6)
plt.show()
```
